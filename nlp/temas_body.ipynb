#### Frecuencia de Palabras

```python
import re
import unidecode

from nltk.probability import FreqDist
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# nltk.download('punkt')
```

```python
stopwords_sp = stopwords.words('spanish')
# agrego palabras poco informativas
stopwords_sp.append('mas')
stopwords_sp.append('ano')
stopwords_sp.append('mes')
stopwords_sp.append('u')
stopwords_sp.append('s')
stopwords_sp.append('casi')
```

```python
def tokenizer(texto):

    alphanumeric = re.sub(r'([^\s\w]|_)+', '', texto).lower()
    no_accents = unidecode.unidecode(alphanumeric)
    
    tockens = word_tokenize(no_accents)
    
    tockens_clean = [tocken for tocken in tockens if tocken not in stopwords_sp and tocken.isalpha()]
    
    terminos = tockens_clean

    return terminos
```

```python
tocken_manual_cambio = [tokenizer(doc) for doc in df_cambio]
```

```python
tocken_manual_infla = [tokenizer(doc) for doc in df_inflacion]
```

### Tokenización y vectorización

Remuevo nros. y carácters especiales antes de "tokenizar"

```python
vectorizer = CountVectorizer(stop_words = stopwords_sp, lowercase = True, strip_accents='unicode')
vectorizer.fit(list(re.sub(r'(\d|\$|\%|\+)', '', doc) for doc in corpus))
```

```python
len(vectorizer.vocabulary_),len(vectorizer.get_feature_names())
```

#### Vectorización

```python
matriz_densa = vectorizer.transform(corpus);
```

```python
matriz_dispersa = matriz_densa.todense()
```

Vectorización con TF-IDF

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
matriz_tfidf = TfidfTransformer().fit_transform(matriz_dispersa);
```

```python
dummies_tfidf = pd.DataFrame(matriz_tfidf.todense(), columns = vectorizer.get_feature_names())
```

Terminos más relevantes por documento

```python
for row in range(dummies_tfidf.shape[0]):
    print(row, dummies_tfidf.iloc[row].sort_values(ascending=False).index[0], dummies_tfidf.iloc[row].sort_values(ascending=False)[0])
```

```python
Relevancia (distinción) de documento en corpus
```

```python
for row in range(dummies_tfidf.shape[0]):
    print('Documento ',row,'\n',
          list(dummies_tfidf.iloc[row].sort_values(ascending=False).index[0:25]), '\n'
          'Relevancia:',
          sum(dummies_tfidf.iloc[row].sort_values(ascending=False)[0:25]) / len(corpus), sep='')
    print('--------------')
```

### Singular Value Decomposition (SVD)

```python
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components = 2) # n_components = Desired dimensionality of output data. Must be strictly less than the number of features.
p = svd.fit_transform(matriz_tfidf)

P = np.around(p,3)
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x = P[:,0], y = P[:,1],
    mode ='markers+text',
    text = ['doc' + str(i) for i in range(len(corpus))],
    marker = dict(
        size = 10,
        color = list(range(29)),
    ),
    textposition="middle left",
    textfont=dict(
        size=10,
        color="Blue"
    )
))
fig.update_layout(title='Tópicos', xaxis_zeroline=True)

fig.show()
```

```python
comp1, comp2 = svd.components_
```
